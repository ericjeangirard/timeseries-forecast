```python
import pandas as pd
import os
import datetime

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dfs = []
for f in os.listdir('/Users/eric/Downloads/'):
    if 'Travel_Times_Daily' in f:
        dfs.append(pd.read_csv('/Users/eric/Downloads/'+f))
df = pd.concat(dfs)
new_col = ['Date',
    'Origin Display Name',
           'Destination Display Name',
    'Daily Mean Travel Time (Seconds)']
df_final = df[new_col].drop_duplicates().sort_values(by='Date')
df_final.to_csv('raw_data.csv', index = False)
```

```python
df = pd.read_csv('raw_data.csv')
df["Date"] = pd.to_datetime(df['Date'], format='%m/%d/%Y')
df = df.sort_values(by='Date').reset_index()
```

```python
df['date_str'] = df.Date.apply(lambda x:str(x)[0:10].replace('-', ''))
df['year'] = df['date_str'].apply(lambda x:x[0:4])
df['month'] = df['date_str'].apply(lambda x:x[4:6])
df['day'] = df['date_str'].apply(lambda x:x[6:8])
df['dow'] = df.Date.apply(lambda x:x.isoweekday())
df
```

```python
df.groupby('year')["Daily Mean Travel Time (Seconds)"].mean().plot()
```
